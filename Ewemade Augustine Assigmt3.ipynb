```python
#Q1

import numpy as np
import matplotlib.pyplot as plt

s = 0.1  # step size
N = 100  # maximum number of iterations
d = 0.01  # tolerance for convergence

x = np.zeros(N)
y = np.zeros(N)

x[0] = 1  # initial x value
y[0] = 1  # initial y value

for i in range(1, N):
    x_inc = -s * (4 * (x[i-1] - 1)**3 - 1)
    y_inc = -s * 2 * (y[i-1] - 2)

    if x_inc**2 + y_inc**2 < d**2:
        break
    else:
        x[i] = x[i-1] + x_inc
        y[i] = y[i-1] + y_inc

# Truncate the arrays to the number of iterations performed
x = x[:i+1]
y = y[:i+1]

# Plotting the progress of the algorithm
plt.figure(figsize=(10, 10))
plt.scatter(x, y, c='blue', marker='o')
plt.plot(x, y, c='blue', linestyle='-', linewidth=1)
plt.scatter([1], [2], c='red', marker='x')  # Global minimum
plt.xlabel('x')
plt.ylabel('y')
plt.title('Steepest Descent Algorithm Progress')
plt.grid()
plt.show()
```

```python
#Q2
#Part (a): Using (1, 1) as the initial point
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 * (x - 2) * (x + 3) + y**2 * (y - 2) * (y + 3)

def grad_f(x, y):
    df_dx = 2 * x * (x - 2) * (x + 3) + x**2 * ((x - 2) + (x + 3))
    df_dy = 2 * y * (y - 2) * (y + 3) + y**2 * ((y - 2) + (y + 3))
    return np.array([df_dx, df_dy])

# Parameters
s = 0.01  # step size
N = 100  # maximum number of iterations
d = 0.01  # tolerance for convergence

# Initial points for part (a)
initial_points = [(1, 1)]

# Execute for part (a) and plot f(x, y) against the number of iterations
for x0, y0 in initial_points:
    x = np.zeros(N)
    y = np.zeros(N)
    f_values = np.zeros(N)

    x[0], y[0] = x0, y0
    f_values[0] = f(x[0], y[0])

    for i in range(1, N):
        grad = grad_f(x[i-1], y[i-1])
        x_inc, y_inc = -s * grad

        if np.linalg.norm(grad) < d:
            break
        else:
            x[i] = x[i-1] + x_inc
            y[i] = y[i-1] + y_inc
            f_values[i] = f(x[i], y[i])

    # Truncate the arrays to the number of iterations performed
    x = x[:i+1]
    y = y[:i+1]
    f_values = f_values[:i+1]

    # Plotting f(x, y) against the number of iterations
    plt.figure(figsize=(10, 5))
    plt.plot(f_values, marker='o', linestyle='-', color='blue')
    plt.xlabel('Iterations')
    plt.ylabel('f(x, y)')
    plt.title('f(x, y) vs. Iterations for Initial Point (1, 1)')
    plt.grid()
    plt.show()
```

```python
#Q2
#Part (b), (c), and (d): Repeat with different initial points
# Initial points for part (b)
initial_points = [(1, -1)]

# Execute for part (b) and plot f(x, y) against the number of iterations
for x0, y0 in initial_points:
    x = np.zeros(N)
    y = np.zeros(N)
    f_values = np.zeros(N)

    x[0], y[0] = x0, y0
    f_values[0] = f(x[0], y[0])

    for i in range(1, N):
        grad = grad_f(x[i-1], y[i-1])
        x_inc, y_inc = -s * grad

        if np.linalg.norm(grad) < d:
            break
        else:
            x[i] = x[i-1] + x_inc
            y[i] = y[i-1] + y_inc
            f_values[i] = f(x[i], y[i])

    # Truncate the arrays to the number of iterations performed
    x = x[:i+1]
    y = y[:i+1]
    f_values = f_values[:i+1]

    # Plotting f(x, y) against the number of iterations
    plt.figure(figsize=(10, 5))
    plt.plot(f_values, marker='o', linestyle='-', color='blue')
    plt.xlabel('Iterations')
    plt.ylabel('f(x, y)')
    plt.title('f(x, y) vs. Iterations for Initial Point (1, -1)')
    plt.grid()
    plt.show()
```
