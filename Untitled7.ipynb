```python
import numpy as np
import random

####################### Parabolic clusters ###########################
##################### Synthetic Data Creation #########################
nData = 2000  # number of points in each cluster
nTrain = 1000
nTest = 2000  # number of test points
std = 0.5  # standard deviation of the distance of points

####################### Nested Training and Testing Loop ###########################
nAlphaSteps = list(np.arange(0.001, 0.01, 0.001)) + list(np.arange(0.01, 0.06, 0.01))  # Alpha range
nEpochSteps = range(1, 7)  # Epoch range

results = []  # To store nSuccess values
for alpha in nAlphaSteps:
    row = []
    for nEpoch in nEpochSteps:
        nSuccess = 0
        for _ in range(50):  # Outer loop for 50 runs
            # Synthetic data creation
            x = np.zeros((2 * nData, 2))  # Pre-allocate for input
            t = np.zeros((2 * nData))  # Target vector
            r = np.random.normal(0, std, 2 * nData)
            theta = 2 * np.pi * np.random.rand(2 * nData)
            h = -6 + 12 * np.random.rand(nData)
            v = 5 + (h ** 2) / 6
            x[:nData, 0] = h + r[:nData] * np.cos(theta[:nData])
            x[:nData, 1] = v + r[:nData] * np.sin(theta[:nData])
            t[:nData] = 0
            h = -5 + 10 * np.random.rand(nData)
            v = 10 + (h ** 2) / 4
            x[nData:, 0] = h + r[nData:] * np.cos(theta[nData:])
            x[nData:, 1] = v + r[nData:] * np.sin(theta[nData:])
            t[nData:] = 1

            # Normalization
            x[:, 0] = 1 + 0.1 * x[:, 0]
            x[:, 1] = 1 + 0.1 * x[:, 1]

            #################### Parameter Initialization #########################
            w11 = 0.5 - np.random.rand()
            w12 = 0.5 - np.random.rand()
            w21 = 0.5 - np.random.rand()
            w22 = 0.5 - np.random.rand()
            b1 = 0.5 - np.random.rand()
            b2 = 0.5 - np.random.rand()
            ww1 = 0.5 - np.random.rand()
            ww2 = 0.5 - np.random.rand()
            bb = 0.5 - np.random.rand()

            a = range(0, 2 * nData)
            sp = random.sample(a, nTrain + nTest)  # Random selection for training and testing

            ########################### Training #####################################
            for epoch in range(nEpoch):
                for i in range(nTrain):
                    y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
                    y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
                    dxx1 = y1 > 0
                    xx1 = y1 * dxx1
                    dxx2 = y2 > 0
                    xx2 = y2 * dxx2  # Rectified activation
                    yy = bb + ww1 * xx1 + ww2 * xx2
                    yy = yy > 0
                    e = t[sp[i]] - yy
                    ww1 += alpha * e * xx1
                    ww2 += alpha * e * xx2
                    bb += alpha * e
                    w11 += alpha * e * ww1 * dxx1 * x[sp[i], 0]
                    w12 += alpha * e * ww1 * dxx1 * x[sp[i], 1]
                    w21 += alpha * e * ww2 * dxx2 * x[sp[i], 0]
                    w22 += alpha * e * ww2 * dxx2 * x[sp[i], 1]
                    b1 += alpha * e * ww1 * dxx1
                    b2 += alpha * e * ww2 * dxx2

            ######################## Testing ######################################
            er = 0
            for i in range(nTrain, nTrain + nTest):
                y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
                y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
                dxx1 = y1 > 0
                xx1 = y1 * dxx1
                dxx2 = y2 > 0
                xx2 = y2 * dxx2  # Rectified activation
                yy = bb + ww1 * xx1 + ww2 * xx2
                yy = yy > 0
                e = abs(t[sp[i]] - yy)
                er += e
            er = er / nTest
            if er < 0.05:
                nSuccess += 1
        row.append(nSuccess)
    results.append(row)

######################## Display Table of nSuccess ###########################
print("Table of nSuccess values:")
print("Rows: Alpha values")
print("Columns: Epoch values")
for i, alpha in enumerate(nAlphaSteps):
    print(f"Alpha: {alpha:.3f} -> {results[i]}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Use the optimal Alpha and nEpoch
alpha = 0.007
nEpoch = 3

# Synthetic data creation (similar setup as before)
# (Keep the data generation and normalization code unchanged)

########################### Training and Testing with Plots ###########################
successful_run = False

for _ in range(50):  # Outer loop for 50 runs
    # Initialize parameters
    w11 = 0.5 - np.random.rand()
    w12 = 0.5 - np.random.rand()
    w21 = 0.5 - np.random.rand()
    w22 = 0.5 - np.random.rand()
    b1 = 0.5 - np.random.rand()
    b2 = 0.5 - np.random.rand()
    ww1 = 0.5 - np.random.rand()
    ww2 = 0.5 - np.random.rand()
    bb = 0.5 - np.random.rand()

    ########################### Training #####################################
    for epoch in range(nEpoch):
        for i in range(nTrain):
            y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
            y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
            dxx1 = y1 > 0
            xx1 = y1 * dxx1
            dxx2 = y2 > 0
            xx2 = y2 * dxx2  # Rectified activation
            yy = bb + ww1 * xx1 + ww2 * xx2
            yy = yy > 0
            e = t[sp[i]] - yy
            ww1 += alpha * e * xx1
            ww2 += alpha * e * xx2
            bb += alpha * e
            w11 += alpha * e * ww1 * dxx1 * x[sp[i], 0]
            w12 += alpha * e * ww1 * dxx1 * x[sp[i], 1]
            w21 += alpha * e * ww2 * dxx2 * x[sp[i], 0]
            w22 += alpha * e * ww2 * dxx2 * x[sp[i], 1]
            b1 += alpha * e * ww1 * dxx1
            b2 += alpha * e * ww2 * dxx2

    ######################## Testing ######################################
    er = 0
    for i in range(nTrain, nTrain + nTest):
        y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
        y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
        dxx1 = y1 > 0
        xx1 = y1 * dxx1
        dxx2 = y2 > 0
        xx2 = y2 * dxx2  # Rectified activation
        yy = bb + ww1 * xx1 + ww2 * xx2
        yy = yy > 0
        e = abs(t[sp[i]] - yy)
        er += e
    er = er / nTest

    if er < 0.05:  # Plotting for successful runs
        successful_run = True
        print(f"Error rate: {er}, creating plots...")

        ########################## Plotting Part ##############################
        plt.figure(figsize=(10, 10))
        plt.scatter(x[0:nData, 0], x[0:nData, 1], label="Cluster 1")
        plt.scatter(x[nData:2 * nData, 0], x[nData:2 * nData, 1], label="Cluster 2")
        X = np.arange(0.25, 1.75, 0.02)
        Y = np.arange(1.25, 2.75, 0.02)
        X, Y = np.meshgrid(X, Y)
        y1 = b1 + w11 * X + w12 * Y
        y2 = b2 + w21 * X + w22 * Y
        dxx1 = y1 > 0
        xx1 = y1 * dxx1
        dxx2 = y2 > 0
        xx2 = y2 * dxx2  # Rectified activation
        yy = bb + ww1 * xx1 + ww2 * xx2
        Z = yy > 0
        plt.scatter(X, Y, c=Z + 1, alpha=0.3)
        plt.legend()
        plt.show()

        plt.figure(figsize=(10, 10))
        f = np.arange(0, nEpoch * nTrain, 1)
        plt.plot(f, [w11] * len(f), label="w11 over epochs")
        plt.plot(f, [w12] * len(f), label="w12 over epochs")
        plt.legend()
        plt.show()

        break

if not successful_run:
    print("No successful runs with error rate < 0.05.")
```
