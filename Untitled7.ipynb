```python
#Q1#
import numpy as np
import random

####################### Parabolic clusters ###########################
##################### Synthetic Data Creation #########################
nData = 2000  # number of points in each cluster
nTrain = 1000
nTest = 2000  # number of test points
std = 0.5  # standard deviation of the distance of points

####################### Nested Training and Testing Loop ###########################
nAlphaSteps = list(np.arange(0.001, 0.01, 0.001)) + list(np.arange(0.01, 0.06, 0.01))  # Alpha range
nEpochSteps = range(1, 7)  # Epoch range

results = []  # To store nSuccess values
for alpha in nAlphaSteps:
    row = []
    for nEpoch in nEpochSteps:
        nSuccess = 0
        for _ in range(50):  # Outer loop for 50 runs
            # Synthetic data creation
            x = np.zeros((2 * nData, 2))  # Pre-allocate for input
            t = np.zeros((2 * nData))  # Target vector
            r = np.random.normal(0, std, 2 * nData)
            theta = 2 * np.pi * np.random.rand(2 * nData)
            h = -6 + 12 * np.random.rand(nData)
            v = 5 + (h ** 2) / 6
            x[:nData, 0] = h + r[:nData] * np.cos(theta[:nData])
            x[:nData, 1] = v + r[:nData] * np.sin(theta[:nData])
            t[:nData] = 0
            h = -5 + 10 * np.random.rand(nData)
            v = 10 + (h ** 2) / 4
            x[nData:, 0] = h + r[nData:] * np.cos(theta[nData:])
            x[nData:, 1] = v + r[nData:] * np.sin(theta[nData:])
            t[nData:] = 1

            # Normalization
            x[:, 0] = 1 + 0.1 * x[:, 0]
            x[:, 1] = 1 + 0.1 * x[:, 1]

            #################### Parameter Initialization #########################
            w11 = 0.5 - np.random.rand()
            w12 = 0.5 - np.random.rand()
            w21 = 0.5 - np.random.rand()
            w22 = 0.5 - np.random.rand()
            b1 = 0.5 - np.random.rand()
            b2 = 0.5 - np.random.rand()
            ww1 = 0.5 - np.random.rand()
            ww2 = 0.5 - np.random.rand()
            bb = 0.5 - np.random.rand()

            a = range(0, 2 * nData)
            sp = random.sample(a, nTrain + nTest)  # Random selection for training and testing

            ########################### Training #####################################
            for epoch in range(nEpoch):
                for i in range(nTrain):
                    y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
                    y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
                    dxx1 = y1 > 0
                    xx1 = y1 * dxx1
                    dxx2 = y2 > 0
                    xx2 = y2 * dxx2  # Rectified activation
                    yy = bb + ww1 * xx1 + ww2 * xx2
                    yy = yy > 0
                    e = t[sp[i]] - yy
                    ww1 += alpha * e * xx1
                    ww2 += alpha * e * xx2
                    bb += alpha * e
                    w11 += alpha * e * ww1 * dxx1 * x[sp[i], 0]
                    w12 += alpha * e * ww1 * dxx1 * x[sp[i], 1]
                    w21 += alpha * e * ww2 * dxx2 * x[sp[i], 0]
                    w22 += alpha * e * ww2 * dxx2 * x[sp[i], 1]
                    b1 += alpha * e * ww1 * dxx1
                    b2 += alpha * e * ww2 * dxx2

            ######################## Testing ######################################
            er = 0
            for i in range(nTrain, nTrain + nTest):
                y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
                y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
                dxx1 = y1 > 0
                xx1 = y1 * dxx1
                dxx2 = y2 > 0
                xx2 = y2 * dxx2  # Rectified activation
                yy = bb + ww1 * xx1 + ww2 * xx2
                yy = yy > 0
                e = abs(t[sp[i]] - yy)
                er += e
            er = er / nTest
            if er < 0.05:
                nSuccess += 1
        row.append(nSuccess)
    results.append(row)

######################## Display Table of nSuccess ###########################
print("Table of nSuccess values:")
print("Rows: Alpha values")
print("Columns: Epoch values")
for i, alpha in enumerate(nAlphaSteps):
    print(f"Alpha: {alpha:.3f} -> {results[i]}")
```

```python
#Q4#
import numpy as np
import random
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid activation function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

##################### Synthetic Data Creation #########################
nData = 2000  # Number of points in each cluster
nTrain = 1000  # Number of training points
nTest = 2000  # Number of test points
std = 0.5  # Standard deviation

# Use the best alpha and nEpoch from the last exercise
alpha = 0.020
nEpoch = 6

nSuccess = 0  # Count successful runs
nSuccess_values = []  # Store success counts for plotting

for run in range(50):  # Outer loop for 50 runs
    ##################### Synthetic Data ########################
    x = np.zeros((2 * nData, 2))  # Input data
    t = np.zeros((2 * nData))  # Target vector
    r = np.random.normal(0, std, 2 * nData)
    theta = 2 * np.pi * np.random.rand(2 * nData)
    h = -6 + 12 * np.random.rand(nData)
    v = 5 + (h ** 2) / 6
    x[:nData, 0] = h + r[:nData] * np.cos(theta[:nData])
    x[:nData, 1] = v + r[:nData] * np.sin(theta[:nData])
    t[:nData] = 0
    h = -5 + 10 * np.random.rand(nData)
    v = 10 + (h ** 2) / 4
    x[nData:, 0] = h + r[nData:] * np.cos(theta[nData:])
    x[nData:, 1] = v + r[nData:] * np.sin(theta[nData:])
    t[nData:] = 1

    # Normalization
    x[:, 0] = 1 + 0.1 * x[:, 0]
    x[:, 1] = 1 + 0.1 * x[:, 1]

    ####################### Parameter Initialization #######################
    w11 = 0.5 - np.random.rand()
    w12 = 0.5 - np.random.rand()
    w13 = 0.5 - np.random.rand()
    w14 = 0.5 - np.random.rand()
    w21 = 0.5 - np.random.rand()
    w22 = 0.5 - np.random.rand()
    w23 = 0.5 - np.random.rand()
    w24 = 0.5 - np.random.rand()
    b1 = 0.5 - np.random.rand()
    b2 = 0.5 - np.random.rand()
    b3 = 0.5 - np.random.rand()
    b4 = 0.5 - np.random.rand()
    ww1 = 0.5 - np.random.rand()
    ww2 = 0.5 - np.random.rand()
    ww3 = 0.5 - np.random.rand()
    ww4 = 0.5 - np.random.rand()
    bb = 0.5 - np.random.rand()

    a = range(0, 2 * nData)
    sp = random.sample(a, nTrain + nTest)  # Randomly select training and testing data

    ########################### Training #####################################
    for epoch in range(nEpoch):
        for i in range(nTrain):
            y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
            y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
            y3 = b3 + w13 * x[sp[i], 0] + w14 * x[sp[i], 1]
            y4 = b4 + w23 * x[sp[i], 0] + w24 * x[sp[i], 1]
            xx1 = sigmoid(y1)
            xx2 = sigmoid(y2)
            xx3 = sigmoid(y3)
            xx4 = sigmoid(y4)
            yy = sigmoid(bb + ww1 * xx1 + ww2 * xx2 + ww3 * xx3 + ww4 * xx4)
            e = t[sp[i]] - yy
            # Update weights and biases using the sigmoid derivative
            delta = e * sigmoid_derivative(yy)
            ww1 += alpha * delta * xx1
            ww2 += alpha * delta * xx2
            ww3 += alpha * delta * xx3
            ww4 += alpha * delta * xx4
            bb += alpha * delta
            delta1 = delta * ww1 * sigmoid_derivative(y1)
            delta2 = delta * ww2 * sigmoid_derivative(y2)
            delta3 = delta * ww3 * sigmoid_derivative(y3)
            delta4 = delta * ww4 * sigmoid_derivative(y4)
            w11 += alpha * delta1 * x[sp[i], 0]
            w12 += alpha * delta1 * x[sp[i], 1]
            w13 += alpha * delta3 * x[sp[i], 0]
            w14 += alpha * delta3 * x[sp[i], 1]
            w21 += alpha * delta2 * x[sp[i], 0]
            w22 += alpha * delta2 * x[sp[i], 1]
            w23 += alpha * delta4 * x[sp[i], 0]
            w24 += alpha * delta4 * x[sp[i], 1]
            b1 += alpha * delta1
            b2 += alpha * delta2
            b3 += alpha * delta3
            b4 += alpha * delta4

    ######################## Testing ######################################
    er = 0
    for i in range(nTrain, nTrain + nTest):
        y1 = b1 + w11 * x[sp[i], 0] + w12 * x[sp[i], 1]
        y2 = b2 + w21 * x[sp[i], 0] + w22 * x[sp[i], 1]
        y3 = b3 + w13 * x[sp[i], 0] + w14 * x[sp[i], 1]
        y4 = b4 + w23 * x[sp[i], 0] + w24 * x[sp[i], 1]
        xx1 = sigmoid(y1)
        xx2 = sigmoid(y2)
        xx3 = sigmoid(y3)
        xx4 = sigmoid(y4)
        yy = sigmoid(bb + ww1 * xx1 + ww2 * xx2 + ww3 * xx3 + ww4 * xx4)
        e = abs(t[sp[i]] - (yy > 0.5))  # Convert sigmoid output to binary
        er += e
    er = er / nTest

    if er < 0.05:
        nSuccess += 1
    nSuccess_values.append(nSuccess)

########################## Plotting Part ##############################
plt.figure(figsize=(10, 6))
plt.plot(range(1, 51), nSuccess_values, label="Sigmoid Activation")
plt.xlabel("Run Number")
plt.ylabel("nSuccess")
plt.title("Performance of Sigmoid Activation")
plt.legend()
plt.show()
```
